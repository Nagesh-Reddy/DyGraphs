```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/N2517/Desktop/developer_survey_2019/survey_results_public.csv', index_col='Respondent')
```

```python
country_grp = df.groupby(['Country'])
```

```python
country_grp
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['PercentKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
python_df.head(20)
```
